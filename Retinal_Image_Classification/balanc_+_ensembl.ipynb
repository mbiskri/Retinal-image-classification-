```python
import os
import shutil
import random

# Paths to the original dataset
original_train_path = '/content/drive/MyDrive/myfinaldataset/train'

# Paths for the new dataset
new_dataset_path = '/content/drive/MyDrive/myfinaldataset1'
new_train_path = os.path.join(new_dataset_path, 'train')

# Parameters
target_size = 2000  # Target 2,000 images per class

def create_directory_structure(output_path, class_dirs):
    """Create directory structure for classes."""
    for class_dir in class_dirs:
        os.makedirs(os.path.join(output_path, class_dir), exist_ok=True)

def downsample_class(class_path, target_size):
    """Downsample a class to the target size."""
    files = os.listdir(class_path)
    random.shuffle(files)
    selected_files = files[:target_size]  # Select the first `target_size` files
    return selected_files

# Create directories for the new dataset
create_directory_structure(new_train_path, os.listdir(original_train_path))

# Downsample and copy images
for class_name in os.listdir(original_train_path):
    class_path = os.path.join(original_train_path, class_name)
    if os.path.isdir(class_path):
        downsampled_files = downsample_class(class_path, target_size)
        downsampled_class_path = os.path.join(new_train_path, class_name)

        # Copy downsampled files to the new directory
        for file_name in downsampled_files:
            shutil.copy(os.path.join(class_path, file_name), os.path.join(downsampled_class_path, file_name))

print("Downsampling complete. Check the new dataset directory.")
```

```python
import os

# Path to the training dataset
train_dir = '/content/drive/MyDrive/myfinaldataset1/train'

# Define train image extensions
image_extensions = ('.png', '.jpg', '.jpeg', '.bmp', '.gif', '.tiff')

# Dictionary to store class-wise image counts
class_image_counts = {}

# Iterate through each class folder
for class_name in os.listdir(train_dir):
    class_path = os.path.join(train_dir, class_name)
    if os.path.isdir(class_path):  # Ensure it's a directory
        image_count = 0
        for file in os.listdir(class_path):
            if file.lower().endswith(image_extensions):
                image_count += 1
        class_image_counts[class_name] = image_count

# Print the number of images per class
for class_name, count in class_image_counts.items():
    print(f"Class '{class_name}': {count} images")
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16, ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import classification_report
```

```python
#  Dataset path
data_dir = '/content/drive/MyDrive/myfinaldataset1'
train_dir = '/content/drive/MyDrive/myfinaldataset1/train'
val_dir = '/content/drive/MyDrive/myfinaldataset1/val'
test_dir = '/content/drive/MyDrive/myfinaldataset1/test'
```

```python
# Data augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    os.path.join(data_dir, 'train'),
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

validation_generator = test_datagen.flow_from_directory(
    os.path.join(data_dir, 'val'),
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)
```

```python
# Function to build model
def build_model(base_model, num_classes):
    x = base_model.output
    x = Flatten()(x)
    x = Dense(1024, activation='relu')(x)
    x = Dropout(0.5)(x)
    predictions = Dense(num_classes, activation='softmax')(x)
    model = Model(inputs=base_model.input, outputs=predictions)
    return model
```

```python
# Load pre-trained models and fine-tune
vgg_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
resnet_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
# Freeze base model layers
for layer in vgg_model.layers:
    layer.trainable = False
for layer in resnet_model.layers:
    layer.trainable = False
```

```python
# Build and compile models
num_classes = len(train_generator.class_indices)
vgg_finetuned = build_model(vgg_model, num_classes)
resnet_finetuned = build_model(resnet_model, num_classes)

vgg_finetuned.compile(optimizer=Adam(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
resnet_finetuned.compile(optimizer=Adam(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Training callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)
```

```python
# Train models
vgg_history = vgg_finetuned.fit(
    train_generator,
    epochs=10,
    validation_data=validation_generator,
    callbacks=[early_stopping, reduce_lr]
)

resnet_history = resnet_finetuned.fit(
    train_generator,
    epochs=30,
    validation_data=validation_generator,
    callbacks=[early_stopping, reduce_lr]
)
```

```python
# Ensemble prediction (averaging)
def ensemble_predictions(models, test_data):
    predictions = [model.predict(test_data) for model in models]
    return np.mean(predictions, axis=0)

test_generator = test_datagen.flow_from_directory(
    os.path.join(data_dir, 'test'),
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)
```

```python
ensemble_preds = ensemble_predictions([vgg_finetuned, resnet_finetuned], test_generator)
final_preds = np.argmax(ensemble_preds, axis=1)
true_classes = test_generator.classes
class_labels = list(test_generator.class_indices.keys())

print(classification_report(true_classes, final_preds, target_names=class_labels))
```

```python
from sklearn.metrics import accuracy_score

# Compute test accuracy
accuracy = accuracy_score(true_classes, final_preds)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```
